### Математическая обработка экспертных оценок

```python
def f1(qinput):
    matrr = []
    matr_res = []
    for i1 in range (6):
        average = sum(qinput[i1])/len(qinput[i1])
        var = np.var(qinput[i1])
        sqrtt = np.sqrt(var)
        minus = average-sqrtt
        plus = average+sqrtt
        cl = [average, var, sqrtt, minus, plus]
        cl_res=[average, plus, minus]
        matrr.append (cl)
        matr_res.append (cl_res)
    return (matrr, matr_res)
    
#Импорт библиотек
import pandas as pd
import numpy as np
import random
from statistics import mean

opros1 = []
for i in range(10):
    expert = []
    for j in range(6):
        expert.append(random.randint(1,10))
    opros1.append(expert)

q = np.array(opros1)
q = q.transpose()

hhh = ['Философия', 'Английский язык', 'ММПР', 'КП', 'Инжиниринг', 'Управленческий учет']
hhh1 = ['Эксперт 1', 'Эксперт 2', 'Эксперт 3', 'Эксперт 4', 'Эксперт 5', 'Эксперт 6', 'Эксперт 7', 'Эксперт 8', 'Эксперт 9', 'Эксперт 10']

print('Массив экспертных оценок, каждому предмету присваивается оценка от 1 до 10, где 1 совсем не важно, 10 очень важно')
df = pd.DataFrame(data=q, index = hhh, columns = hhh1)
df
```

```python
m1 = f1(q)

hhh2 = ['Мат. ожидание', 'Дисперсия', 'Среднеквадратичное отклонение', 'Мат ож. - σ', 'Мат ож. + σ']

df1 = pd.DataFrame(data=m1[0], index = hhh, columns = hhh2)
df1
```

```python
# функция записи в массив обработки бинарного опроса
def BA(Qi, ab):
    a = ab[0] #первый предмет из бинарного опроса
    b = ab[1] #второй предмет из бинарного опроса
    if Qi == hhh[a]: # если результат опроса равен первый предмет из бинарного опроса
        M_U[ (a, b) ] = 1 
        M_U[ (b, a) ] = 0
    else:
        M_U[ (a, b) ] = 0
        M_U[ (b, a) ] = 1   

import pandas as pd 
import numpy as np
from openpyxl import load_workbook
#Формирование исходных данных из файла опроса
file = 'oz.xlsx'
xl = pd.ExcelFile(file)
df1 = xl.parse('Sheet1')
q = np.array(df1)
ML= np.array(q[0:52, 3]) #массив из поля Почта
Q0_15 = np.array(q[0:52, 12:27]) #массив из полей Выберите один из двух предметов ... Выберите один из двух предметов15
MB = np.column_stack([ML, Q0_15]) # массив бинарного опроса объединение массивов ML и Q0_15
#print(MB) 

hhh = ['Философия', 'Английский язык', 'Математические методы принятия решений', 'Компьютерный практикум', 'Инжиниринг бизнеса', 'Управленческий учет']
lbl = ['Философия', 'Английский', 'ММПР', 'КП', 'Инжиниринг', 'Учет'] # Для подписей к таблице

# матрица M_K Кемени
print ('Формируем матрицу Кемени')
M_K = np.zeros(( len(hhh), len(hhh) )) # размерность матрицы из списка предметов hhh
for j in  range( len(hhh) ): #под диагональю 1 выше диагонали 0
    for j1 in range ( j+1 ):
        M_K [  ( j, j1) ] = 1
print(M_K)
CF = []
for i in MB:
    print('матрица опроса пользователя: ',i[0])
    M_U = np.zeros(( len(hhh), len(hhh) )) # размерность матрицы из списка предметов hhh
    for j in  range( len(hhh) ): # Диагональ заполняем 1
        M_U [ (j, j) ] = 1
    #анализ матрицы по 15 вопросам
    for n in range(1,16):
        Qi = i[n]
        if n ==1: ab = (0,1) #Выберите один из двух предметов (выбор между hhh[0] = Философией - 1  или  hhh[1]=Английским - 0)
        if n ==2: ab = (0,2) #Выберите один из двух предметов2 (выбор между hhh[0] = Философией - 1  или  hhh[2]=ММПР - 0)
        if n ==3: ab = (0,3) #Выберите один из двух предметов3 (выбор между hhh[0] = Философией - 1  или  hhh[3]=КП - 0)
        if n ==4: ab = (0,4) #Выберите один из двух предметов4 (выбор между hhh[0] = Философией - 1  или  hhh[4]=Инжиниринг - 0)
        if n ==5: ab = (0,5) #Выберите один из двух предметов5 (выбор между hhh[0] = Философией - 1  или  hhh[5]=Управленческий учет - 0)
        if n ==6: ab = (1,2) #Выберите один из двух предметов6 (выбор между hhh[1] = Английским - 1  или  hhh[2]=ММПР - 0)
        if n ==7: ab = (1,3) #Выберите один из двух предметов7 (выбор между hhh[1] = Английским - 1  или  hhh[3]=КП - 0)
        if n ==8: ab = (1,4) #Выберите один из двух предметов8 (выбор между hhh[1] = Английским - 1  или  hhh[4]=Инжиниринг - 0)
        if n ==9: ab = (1,5) #Выберите один из двух предметов9 (выбор между hhh[1] = Английским - 1  или  hhh[5]=Управленческий учет - 0)
        if n ==10: ab = (2,3) #Выберите один из двух предметов10 (выбор между hhh[2] = ММПР - 1  или  hhh[3]=КП - 0)
        if n ==11: ab = (2,4) #Выберите один из двух предметов11 (выбор между hhh[2] = ММПР - 1  или  hhh[4]=Инжиниринг - 0)
        if n ==12: ab = (2,5) #Выберите один из двух предметов12 (выбор между hhh[2] = ММПР - 1  или  hhh[5]=Управленческий учет - 0)
        if n ==13: ab = (3,4) #Выберите один из двух предметов13 (выбор между hhh[3] = КП - 1  или  hhh[4]=Инжиниринг - 0)
        if n ==14: ab = (3,5) #Выберите один из двух предметов13 (выбор между hhh[3] = КП - 1  или  hhh[5]=Управленческий учет - 0)
        if n ==15: ab = (4,5) #Выберите один из двух предметов13 (выбор между hhh[4] = Инжиниринг - 1  или  hhh[5]=Управленческий учет - 0)
        BA(Qi, ab) #вызов функции записи в массив M_U после анализа

    df = pd.DataFrame(data=M_U, index = lbl, columns = lbl)
    SSQR = np.sum((M_U - M_K)**2) #Расчет СУММКВРАЗН по пользователю и Кемени
    print('Сумма квадратов разницы: ', SSQR )
    CF.append(SSQR)
    print(df)
print('Целевая функция: ', sum(CF) )
```
